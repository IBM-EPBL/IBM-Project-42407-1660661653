# **Outliers Replacement**

```python
sns.boxplot(df.Age)
```

```python
q1=df.Age.quantile(0.25)
q3=df.Age.quantile(0.75)
```

```python
IQR = q3-q1
```

```python
upper_limit=q3 + 1.5 * IQR
```

```python
upper_limit
```

```python
df.Age=np.where(df.Age>upper_limit,44,df.Age) #Median=44.000000	
```

```python
sns.boxplot(df.Age)
```
